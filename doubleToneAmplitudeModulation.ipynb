```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
fc = 1000  # Carrier frequency (Hz)
fm = 100   # Modulating frequency (Hz)
Ac = 1     # Carrier amplitude
Am = 0.5   # Modulating amplitude
m1 = 0.5   # Modulation index

# Time array
t = np.linspace(0, 1, 1000)  # 1 second duration

# Carrier signal
carrier_signal = Ac * np.cos(2 * np.pi * fc * t)

# Modulating signal
modulating_signal = Am * np.cos(2 * np.pi * fm * t)

# Double-tone amplitude modulated signal
DTAM_signal = (1 + m1 * np.cos(2 * np.pi * fm * t)) * Ac * np.cos(2 * np.pi * fc * t)

# Plotting
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(t, carrier_signal, 'r')
plt.title('Carrier Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(t, modulating_signal, 'g')
plt.title('Modulating Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t, DTAM_signal, 'b')
plt.title('Double-tone Amplitude Modulated Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```
